#### 2. How many different Profiles do students do into ? 

##### Note : A lot of overlapping is here, example : Sde, Software Developer, Software Engineer all are same

```python
total_different_profiles = 0
set_of_profiles = set()
for i in range(len(df)):
    profile_here = df.loc[i, "JobProfile"]
    try:
        for x in profile_here.split(','):
            set_of_profiles.add(x)
    except Exception:
        pass

pprint(f"Total Different Profiles are {len(set_of_profiles)}")
pprint("All the different profiles are here")
pprint(set_of_profiles)
```

#### 3. How many different places in India are students placed ? 
##### Note : We also have overlapping places here : Like Bangalore, Bengaluru etc

```python
total_diff_places = 0
set_of_places = set()
for i in range(len(df)):
    place_here = df.loc[i, "JobLocation"]
    try:
        for x in place_here.split(','):
            set_of_places.add(x)
    except Exception :
        pass
pprint(f"Total Different Job Locations are {len(set_of_places)}")
pprint("All the different Job Locations are here")
pprint(set_of_places)
```

#### 4. A lot of myth in college is around this "People in core branches are less placed." <br />
#### a) My hold in this is that there is direct correlation between the `number of companies` coming and `number of people getting placed`.

```python
#These are the count of companies that allow 'variable_name' branch to sit. 
#Example if cs = 50, then 50 companies allowed "CSE" to sit for the test.
cs = 0; it = 0; ece = 0; civil = 0; mech = 0; pie = 0; ee = 0
total_diff_branches = 0
set_of_branches = set()
for i in range(len(df)):
    branches_here = df.loc[i, "EligibleBranches"]
    try:
        for x in branches_here.split(','):
            branch = x.strip()
            set_of_branches.add(branch)
            if(branch == "cs"):
                cs += 1
            elif(branch == "it"):
                it += 1
            elif(branch == "pie"):
                pie += 1
            elif(branch == "mech"):
                mech += 1
            elif(branch == "ece"):
                ece += 1
            elif(branch == "ee"):
                ee += 1
            else:
                civil += 1
    except:
        pass
pprint(set_of_branches)
pprint(f"Total Companies that allowed CSE : {cs}")
pprint(f"Total Companies that allowed IT : {it}")
pprint(f"Total Companies that allowed ECE : {ece}")
pprint(f"Total Companies that allowed EE : {ee}")
pprint(f"Total Companies that allowed CIVIL : {civil}")
pprint(f"Total Companies that allowed MECH : {mech}")
pprint(f"Total Companies that allowed PIE : {pie}")
```

#### 4. Lets talk about the CGPA crtiteria here. Usually seniors tell juniors to maintain a good enough 7 an you will be good to go. Let's check if that thing follows.

```python
total_diff_cgpa = []
s = set()
for i in range(len(df)):
    cgpa_here = df.loc[i, "MinCgpa"]
    if(pd.isnull(cgpa_here)):
        cgpa_here = str(0)
    cgpa_here = cgpa_here.split(',')[0].split(' ')
    cgpa_here = cgpa_here[0]
    # Note some companies have no cgpa criteria. 
        ## In here there are 2 cases.
            ### a) That donot allow active backlogs
            ### b) That allow active backlogs
        # While gathering this data, I have found that type(a) companies are much great
        # er than type(b) companies. So please don't have any active backlog while sitting for placements.     
    if(cgpa_here == "all" or cgpa_here == "backlogs" or cgpa_here == "na" or cgpa_here == "nb" or cgpa_here == "pwd"):
        cgpa_here = str(0)
    s.add(cgpa_here)
    total_diff_cgpa.append(np.float(cgpa_here))


    # Sort this CGPA list
total_diff_cgpa = sorted(total_diff_cgpa)
total_diff_cgpa

assert(len(total_diff_cgpa) == len(df))

pprint(f"The Maximum CGPA required is {total_diff_cgpa[-1]}")
pprint("Note : 0 means that there is no requirement on CGPA but it maybe that you have no active backlogs")
pprint(f"The Minimum CGPA required is {total_diff_cgpa[0]}")

# Lets see some Percentiles
pprint(f"Note : Here it should be provided that you are in EligibleBranches")
pprint(f"If you want to sit for 100% of companies. {total_diff_cgpa[-1]}")
pprint(f"If you want to sit for 75% of companies. {total_diff_cgpa[int(0.75*len(total_diff_cgpa))]}")
pprint(f"If you want to sit for 50% of companies. {total_diff_cgpa[int(0.50*len(total_diff_cgpa))]}")
pprint(f"If you want to sit for 40% of companies. {total_diff_cgpa[int(0.40*len(total_diff_cgpa))]}")
```
